# Lab 2 of Exercise 2.B, working
# with datetime functions

```python
import datetime
```

# Today

```python
today = datetime.datetime.now()
```

```python
print("Today is: " + today.strftime("%A, %B %d, %Y"))
```

# My birthday

```python
birthday = datetime.date(2027,2,10)
```

```python
print("My birthday is: " + birthday.strftime("%A, %B, %d, %Y"))
```

# Next 90 Days

```python
ninety_d = today + datetime.timedelta(days=90)
```

```python
print("In 90 days: " + ninety_d.strftime("%A, %B, %d, %Y"))
```

# Time for Dinner

```python
dinner_time = datetime.time(19,30)
```

```python
print("Dinner time is: " + dinner_time.strftime("%I:%M %p"))
```
