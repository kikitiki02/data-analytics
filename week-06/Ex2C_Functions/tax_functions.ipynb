```python
def get_soc_sec_tax(gross_pay):
    return gross_pay * 0.062
```

```python
def get_medicare_tax(gross_pay):
    return gross_pay * 0.0145
```

```python
# Withholding Code Tax Rate
# 0 23%
# 1 21%
# 2 19.5%
# 3 18.5%
#4+ 18% less 0.5% for each withholding over 4
```

```python
def get_federal_tax(gross_pay, w_code):
    if w_code == 0:
        t_r = 0.23
    elif w_code == 1:
        t_r = 0.21
    elif w_code == 2:
        t_r = 0.195
    elif w_code == 3:
        t_r = 0.185
    else:
        t_r = 0.18
    return gross_pay * t_r
    
```

```python
# Person1: gross pay $750 withholding code 0
# Person2: gross pay $1550 withholding code 2
# Person3: gross pay $1100 withholding code 5
```

```python
gross_pay = 750
w_code = 0 
print("Person 1")
print("Social Security Tax:", get_soc_sec_tax(gross_pay))
print("Medicare Tax:", get_medicare_tax(gross_pay))
print("Federal Tax:", get_federal_tax(gross_pay, w_code))
```

```python
gross_pay = 1550
w_code = 2 
print("Person 2")
print("Social Security Tax:", get_soc_sec_tax(gross_pay))
print("Medicare Tax:", get_medicare_tax(gross_pay))
print("Federal Tax:", get_federal_tax(gross_pay, w_code))
```
