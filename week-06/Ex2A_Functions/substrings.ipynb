## Lab 1. Working with names – parsing strings

```python
name = "Kiki Marsh"
```

```python
def parse_and_display(name):
    for i in range(len(name)):
        if name[i] == ' ':
            first = name[:i]
            last = name[i+1:]
            print(f"Full Name: {name}")
            print(f"First Name: {first}")
            print(f"Last Name: {last}")
```

```python
parse_and_display(name)
```

```python
test_list = ['Lipi Olive', 'Hello Kitty', 'Toji Fushiguro']
```

```python
parse_and_display(test_list[0]) # output for Lipi Olive
```

```python
parse_and_display(test_list[1]) # output for Hello Kitty
```

```python
parse_and_display(test_list[2]) # output for Toji Fushiguro
```

```python
parse_and_display('Kiersten Marsh')
```
